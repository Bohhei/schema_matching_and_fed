```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
df_numeric = pd.read_csv("openml_15ds_attributes_numeric_matching.csv")#df[[df.columns[1], df.columns[5], df.columns[2], df.columns[6], df.columns[-1], df.columns[0], df.columns[4]]]
df_numeric.columns
```

```python
df_num = pd.read_csv("openml_15ds_attributes_numeric_index.csv")
df_num.head()
```

```python
def find_index(df_n, at, id_):
    ind = []
    for i in df_n.index[df_n['attr_name'] == at]:
        for j in df_n.index[df_n['dataset_openml_id'] == id_]:
            if i==j:
                ind.append(i)
    if len(ind)>1:
        print('len>1', at, id_)
        return ind
    else:
        return ind[0]
```

```python
constr_num = []
name1=[]
name2=[]

for i in df_numeric.index:
    df_ = df_numeric.loc[i]
    attr1 = df_[2][1:-1]
    attr2 = df_[6][1:-1]
    d1 = df_[0]
    d2 = df_[4]
    d_name1 = df_[1]
    d_name2 = df_[5]
    idx1 = find_index(df_num, attr1, d1)
    idx2 = find_index(df_num, attr2, d2)
    constr1 = df_num.loc[idx1][df_num.columns[4:]]
    constr2 = df_num.loc[idx2][df_num.columns[4:]]
    constr = list(constr1.values) + list(constr2.values) + [df_[-3]]
    constr_num.append(constr)
    name1.append(d_name1 + str(d1))
    name2.append(d_name2 + str(d2))

    
```

```python
constr_num[0]
```

```python
new_df_num = pd.DataFrame()
new_df_num['dataset1_name'] = name1
new_df_num['dataset2_name'] = name2
new_df_num['attr1_name'] = df_numeric[df_numeric.columns[2]].values
new_df_num['attr2_name'] = df_numeric[df_numeric.columns[6]].values
new_df_num['attribute_match'] = df_numeric[df_numeric.columns[-1]].values
new_df_num['constraints'] = constr_num
new_df_num
```

```python
new_df_num.to_pickle('numeric2.p')
```

```python
# 'dataset1_name', 'dataset2_name', 'attr1_name', 'attr2_name', 'attributes_match' for SMAT
# 
df_nominal = pd.read_csv("openml_15ds_attributes_nominal_matching.csv")#df[['dataset1_name', 'dataset2_name', 'attr1_name', 'attr2_name', 'attributes_match', 'dataset1_openml_id', 'dataset2_openml_id']]
df_nominal.columns
```
